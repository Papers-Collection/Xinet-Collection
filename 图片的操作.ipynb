## 图片的对齐方式

使用 html 中的 `div` 标签，它有个对齐属性 `align`，可以指定为 `left`、`center` 和 `right` 以实现*左对齐*、*居中对齐*和*右对齐*。这种方式 div 标签之间不能放 Markdown 图片链接语法，而只能是 html 的语法格式，因此还要配合 `<img>` 标签使用。

```md
<div align="center">
<img src="http://static.zybuluo.com/xinet/dk25o6spbuvf486y2kh8b7oq/%E9%9B%B7%E5%A7%86.PNG" width="75%" height="25%" />
    雷姆
</div>

------------------
<div align="left">
<img src="http://static.zybuluo.com/xinet/dk25o6spbuvf486y2kh8b7oq/%E9%9B%B7%E5%A7%86.PNG" />
    雷姆
</div>

-----------------------
<div align="right">
<img src="http://static.zybuluo.com/xinet/dk25o6spbuvf486y2kh8b7oq/%E9%9B%B7%E5%A7%86.PNG" />
    雷姆
</div>
```
